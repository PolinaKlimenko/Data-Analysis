```python
df = pd.read_csv('bank-full.csv', sep=';')
df.head()
```

```python
# Имена переменных
columns = ['age', 'job', 'marital', 'education', 'default',
           'balance', 'housing', 'loan', 'contact', 'day',
           'month', 'duration', 'campaign', 'pdays', 'previous', 'poutcome']
# Преобразуем категориальные переменные
job = {'admin.':0, 'blue-collar':1, 'entrepreneur':2, 'housemaid':3, 'management':4, 'retired':5, 'self-employed':6,
           'services':7, 'student':8, 'technician':9, 'unemployed':10, 'unknown':11}
df['job'] = df['job'].map(job)

df['marital'] = df['marital'].map({'divorced':0, 'married':1, 'single':2, 'unknown':3})

df['education'] = df['education'].map({'primary':0, 'secondary':1, 'tertiary':2, 'unknown':3})

df['default'] = df['default'].map({'no':0, 'yes':1, 'unknown':2})

df['housing'] = df['housing'].map({'no':0, 'yes':1, 'unknown':2})

df['loan'] = df['loan'].map({'no':0, 'yes':1, 'unknown':2})

df['contact'] = df['contact'].map( {'cellular':0, 'telephone':1, 'unknown':2})

df['month'] = df['month'].map({'jan':0, 'feb':1, 'mar':2, 'apr':3, 'may':4, 'jun':5, 'jul':6, 'aug':7, 'sep':8, 'oct':9, 'nov':10, 'dec':11})

df['poutcome'] = df['poutcome'].map({'failure':0, 'nonexistent':1, 'success':2, 'other':3, 'unknown':4})
# Кодируем отклик значениями 0 и 1
df['y'] = df['y'].map({'no': 0, 'yes': 1})
```

```python
df.head()
```

```python
df.describe()
```

```python
y = df['y']
X = df.drop(columns=['y'])
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split  
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from xgboost import XGBClassifier

# from sklearn.metrics import classification_report

from sklearn import metrics
```

```python
# Инициализируем модель
model = XGBClassifier(base_score=0.5,
                      n_estimators=100,
                      max_depth=6,
                      seed=13,
                      objective='binary:logistic',
                      colsample_bylevel=1,
                      colsample_bytree=0.8,
                      subsample=0.66,
                      gamma=0.001,
                      learning_rate=0.01,
                      max_delta_step=0,
                      min_child_weight=1,
                      nthread=-1,
                      reg_alpha=0,
                      reg_lambda=1,
                      eval_metric='logloss')
```

```python
# Определим тестируемые значения гиперпараметров
params_set = {
    'n_estimators': [250, 500],
    'max_depth': [3],
    'learning_rate': [0.01, 0.1],
    'reg_lambda': [0.01, 0.1],
    'reg_alpha': [0.01, 0.1],
}
```

```python
from sklearn.model_selection import GridSearchCV
# Проведем 5-fold кросс-валидацию
grid_CV = GridSearchCV(estimator=model,
                       param_grid=params_set,
                       scoring='balanced_accuracy', 
                       cv=5,
                       n_jobs=-1,
                       error_score='raise',
                       pre_dispatch='2*n_jobs',
                       refit=True)

# Обучим на тренировочной выборке
grid_CV.fit(X_train, y_train)
```

```python
# Выберем лучшую модель
best_model = grid_CV.best_estimator_
```

```python
# Строим предсказания модели
y_pred_train = best_model.predict(X_train)
y_pred_test = best_model.predict(X_test)

print(metrics.classification_report(y_pred_test, y_test))
```

```python
# Оценим качество классификации первой модели
metrics.balanced_accuracy_score(y_test, y_pred_test)
```

```python
from sklearn.metrics import accuracy_score

print("Accuracy on train set is ", accuracy_score(y_train,y_pred_train))
print("Accuracy on test set is ", accuracy_score(y_test, y_pred_test))
```

```python
# Посмотрим на значимость признаков
pd.DataFrame({'features': X.columns, 'importance': best_model.feature_importances_}).sort_values('importance',
                                                                                                   ascending=False)
```
